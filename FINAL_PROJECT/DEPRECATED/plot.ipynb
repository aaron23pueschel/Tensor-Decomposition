```python
import matplotlib.pyplot as plt
import numpy as np

# Kernel names
kernels = ['ttm_1', 'ttm_2', 'ttm_3']

# DRAM read throughput averages
read_throughput_avg = {
    'ttm_1': 3.9968,
    'ttm_2': 50.353,
    'ttm_3': 18.226
}

# Plotting
index = np.arange(len(kernels))
read_values_avg = [read_throughput_avg[kernel] for kernel in kernels]

fig, ax = plt.subplots(figsize=(8, 6))

read_bars_avg = ax.bar(kernels, read_values_avg, color='b')

ax.set_xlabel('Kernels')
ax.set_ylabel('Read Throughput (GB/s)')
ax.set_title('DRAM Read Throughput (Averages) for Kernels')

def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(read_bars_avg)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Peak Performance and Kernel Throughputs
peak_performance = 10690000000
ttm1_throughput = 480367335
ttm2_throughput = 223889016
ttm3_throughput = 29749588

# Calculate relative performance in percentage
ttm1_percent = (ttm1_throughput / peak_performance) * 100
ttm2_percent = (ttm2_throughput / peak_performance) * 100
ttm3_percent = (ttm3_throughput / peak_performance) * 100

# Names of the kernels
kernels = ['Peak', 'ttm1', 'ttm2', 'ttm3']

# Throughput values
throughputs = [100, ttm1_percent, ttm2_percent, ttm3_percent]

# Colors for bars
colors = ['gray', 'blue', 'green', 'red']

# Create a bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(kernels, throughputs, color=colors)
peak_vals = ['(10.69 Tb/s)', '(480.37 Gb/s)', '(223.89 Gb/s)', '(29.75 Gb/s)']
# Add text annotations
for i,bar in enumerate(bars):
    height = bar.get_height()
    plt.annotate(f'{height:.2f}% {peak_vals[i]}', 
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3), 
                 textcoords="offset points",
                 ha='center', va='bottom')

# Set y-axis label and title
plt.ylabel('Relative Throughput (%)')
plt.title('Kernel Flops VS peak performance')

# Add a reference line for peak performance
plt.axhline(y=100, color='k', linestyle='--', label='Peak Performance')

# Add legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
kernels = ['ttm_1', 'ttm_2', 'ttm_3', 'maxwell_sgemm_128x128_nn']
metrics = ['DRAM Read', 'DRAM Write']
avg_throughput = [[6.40855, 10.879], [53.125, 3.437], [16.537, 1.94965], [57.437, 54.442]]

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.35
index = np.arange(len(kernels))

for i, metric in enumerate(metrics):
    avg_values = [entry[i] for entry in avg_throughput]
    ax.bar(index + i * bar_width, avg_values, bar_width, label=f'{metric}')

ax.set_xlabel('Kernel')
ax.set_ylabel('Average Throughput (GB/s)')
ax.set_title('Average DRAM Throughput for Different Kernels')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(kernels, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```
